# Удаляем строки с неизвестными регионами, так как не понятно, куда отнести информацию, содержащуюся внутри этих строк

```python
all_objects_1 = all_objects.dropna(subset=['Region'])
```

```python
all_objects_1.isnull().sum()
```

# Считаем количество объектов в каждом регионе

```python
region_count = all_objects_1.groupby('Region').agg({'Region_Code': 'count'}) \
    .rename(columns={'Region_Code' : 'Numbers'}).sort_values('Numbers', ascending=False)
```

```python
region_count.head()
```

```python
rcParams['figure.figsize'] = 20,20
region_count.plot(kind='barh')
plt.show()
```

# Считаем среднюю цену за квадратный метр по регионам

```python
mean_price_for_region = all_objects_1.groupby('Region').agg({'Avg_Price_Per_Square': 'mean'}) \
    .sort_values('Avg_Price_Per_Square', ascending=False)
```

```python
mean_price_for_region.head()
```

```python
rcParams['figure.figsize'] = 20,20
mean_price_for_region.plot(kind='barh')
plt.show()
```

# Выявляем среднюю распроданность по регионам

```python
sold_out = all_objects_1.groupby('Region').agg({'Sold_Out': 'mean'}) \
    .sort_values('Sold_Out', ascending=False)
```

```python
sold_out.head()
```

```python
rcParams['figure.figsize'] = 20,20
sold_out.plot(kind='barh')
plt.show()
```

# Выяснилось, что в среднем строящиеся объекты распроданы на 45,2%

```python
sold_out.mean()
```

# Далее считаем суммарную площадь строящегося жилья по регионам

```python
living_space_sum = all_objects_1.groupby('Region').agg({'Living_Space': 'sum'}) \
    .sort_values('Living_Space', ascending=False)
```

```python
living_space_sum.head()
```

```python
rcParams['figure.figsize'] = 20,20
living_space_sum.plot(kind='barh')
plt.show()
```

# Считаем общее количество квартир по регионам

```python
number_of_flats = all_objects_1.groupby('Region').agg({'Number_of_Flats': 'sum'}) \
    .sort_values('Number_of_Flats', ascending=False)
```

```python
number_of_flats.head()
```

```python
rcParams['figure.figsize'] = 20,20
number_of_flats.plot(kind='barh')
plt.show()
```
